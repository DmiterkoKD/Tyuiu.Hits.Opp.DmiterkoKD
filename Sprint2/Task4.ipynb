```polyglot-notebook
public class Employee
{
    public string Name {get; set;}
    public string Position{get; set;}
    public double Salary{get; set;}
    public Employee(string name, string position, double salary)
    {
        Name = name;
        Position = position;
        Salary = salary;
    }
    public virtual void PrintInfo()
    {
        Console.WriteLine($"Сотрудник: {Name}, Должность: {Position}, Зарплата: {Salary}");
    }
}
public class Manager: Employee
{
    public string Department{get; set;}
    public Manager(string name, string position, double salary, string department): base(name, position, salary)
    {
        Department = department;
    }
    public override void PrintInfo()
    {
        base.PrintInfo();
        Console.WriteLine($"Отдел: {Department}");
    }
}
public class ProjectManager : Manager
{
    public string ProjectName{get; set;}
    public ProjectManager(string name, string position, double salary, string department, string projectName)
        : base(name, position, salary, department)
    {
        ProjectName = projectName;
    }

    public override void PrintInfo()
    {
        base.PrintInfo();
        Console.WriteLine($"Проект: {ProjectName}");
    }
}
public interface IReportable
{
    string GenerateReport();
}
public interface ITeamLeader
{
    string[] GetTeamMembers();
}
public class ProjectManagerWithInterfaces : ProjectManager, IReportable, ITeamLeader
{
    private string[] teamMembers = { "Анна Иванова", "Петр Сидоров", "Мария Петрова" };

    public ProjectManagerWithInterfaces(string name, string position, double salary, string department, string projectName): base(name, position, salary, department, projectName)
    {
    }
    public string GenerateReport()
    {
        return $"Отчет по проекту {ProjectName} " + $"Руководитель: {Name}";
    }

    public string[] GetTeamMembers()
    {
        return teamMembers;
    }
}
Employee em = new Employee("Петр", "Бухгалтер", 30000);
em.PrintInfo();
Console.WriteLine();
Manager m = new Manager("Сергей", "Менеджер", 50000, "Торговля");
m.PrintInfo();
Console.WriteLine();
ProjectManager pm = new ProjectManager("Екатерина", "Менеджер", 50000, "Маркетинг", "Йогурт Фру");
pm.PrintInfo();
Console.WriteLine();
ProjectManagerWithInterfaces lll = new ProjectManagerWithInterfaces("Ольга Морозова", "Проектный менеджер", 130000.0, "Разработка", "Мобильное приложение");
lll.PrintInfo();
Console.WriteLine();
Console.WriteLine(lll.GenerateReport());
Console.WriteLine();

string[] members = lll.GetTeamMembers();
foreach (string member in members)
{
    Console.WriteLine($"- {member}");
}

```
